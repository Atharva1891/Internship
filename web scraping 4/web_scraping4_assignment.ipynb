### 10-  Details of Datasets from UCI machine learning repositories. Url = https://archive.ics.uci.edu/ You have to find the following details:
#### A) Dataset name
#### B) Data type
#### C) Task
#### D) Attribute type
#### E) No of instances
#### F) No of attribute
#### G) Year

```python
#connect to the web driver
driver = webdriver.Chrome(r"chromedriver.exe")

#Get url in web driver first
url = 'https://archive.ics.uci.edu/'
driver.get(url)
```

```python
view_all=driver.find_element(By.XPATH,"/html/body/table[1]/tbody/tr/td[2]/span[2]/a").get_attribute("href")
print(view_all)
driver.get(view_all)
```

```python
dataset_name=[]
data_type=[]
task=[]
attribute_type=[]
number_of_attributes=[]
number_of_instances=[]
year=[]

#dataset name
for names in driver.find_elements(By.XPATH,"/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[1]/table/tbody/tr/td[2]/p/b/a"):
    dataset_name.append(names.text)
#data type
for dtype in driver.find_elements(By.XPATH,"/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[2]/p"):
    data_type.append(dtype.text)
#task
for t in driver.find_elements(By.XPATH,"/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[3]"):
    task.append(t.text)
#type of attribute    
for attribute in driver.find_elements(By.XPATH,"/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[4]/p"):
    attribute_type.append(attribute.text)
#number of attributes
for attribute_count in driver.find_elements(By.XPATH,"/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[6]"):
    number_of_attributes.append(attribute_count.text)
#number_of_instances
for instances in driver.find_elements(By.XPATH,"/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[5]"):
    number_of_instances.append(instances.text)
#year
for y in driver.find_elements(By.XPATH,"/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[7]"):
    year.append(y.text)

print(dataset_name[0:2])
print(data_type[0:2])
print(task[0:2])
print(attribute_type[0:2])
print(number_of_attributes[0:2])
print(number_of_instances[0:2])
print(year[0:2])
```

```python
print(len(dataset_name))
print(len(data_type))
print(len(task))
print(len(attribute_type))
print(len(number_of_attributes))
print(len(number_of_instances))
print(len(year))
```

```python
df=pd.DataFrame({})
df["Dataset Name"]=dataset_name
df["Data Type"]=data_type[1:] #slicing list at 1 as it includes the heading of column
df["Task"]=task[1:]
df["Attribute type"]=attribute_type[1:]
df["No of instances"]=number_of_instances[1:]
df["No of attribute"]=number_of_attributes[1:]
df["Year"]=year[1:]
df.head()
```
